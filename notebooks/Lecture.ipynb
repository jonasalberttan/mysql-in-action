<h4 id="CHARACTER-SET"><span style="text-decoration: underline;">CHARACTER SET</span></h4>
<p>A MySQL character set is a set of characters that are legal in a string. The default character set is Latin1 or depending on the setup during the installation of MySQL. Let's execute these lines of code to see what is the default character set and collation after creating the <span style="font-size: 12pt;"><strong>test</strong></span>&nbsp;database.</p>

```mysql
create database if not exists test;
use test;
select @@character_set_database, @@collation_database;
```

<h4 id="COLLATION"><span style="text-decoration: underline;">COLLATION</span></h4>
<p>A MySQL collation is a set of rules used to compare characters in a particular character set. Each character set in MySQL can have more than one collation, and has, at least, one default collation. Two character sets cannot have the same collation.</p>

```mysql
show collation like 'utf8_%';
```

<p>Let's create a database and specify a character set and collation.</p>

```mysql
drop database if exists test;
create database test character set utf8 collate utf8_unicode_ci;
```

<h2 id="4.-Derived-Table" style="font-family: 'helvetica neue' , 'helvetica' , 'arial' , sans-serif; color: #000000;">4. <span style="text-decoration: underline;">Derived Table</span></h2>
<p><span style="font-size: 12pt;">The derived table persists only for the duration of the query, in contrast with a temporary table, which exists for the entire session. Later we will learn about the temporary table.</span></p>
<h4 id="Derived-Table-in-Practice">Derived Table in Practice</h4>
<p style="padding-left: 40px;"><span style="font-size: 12pt;">1. If the query results are only needed once, you might instead use a derived table.</span></p>
<p style="padding-left: 40px;"><span style="font-size: 12pt;">2. </span><span style="font-family: , , 'segoe ui' , 'roboto' , 'oxygen' , 'ubuntu' , 'cantarell' , 'open sans' , 'helvetica neue' , sans-serif; font-size: medium;">Avoid derived tables.&nbsp; If there is another way to write the query it will be faster in most cases. In many cases, even a separate temporary table will be faster as you can add proper indexes to the table in this case.</span></p>
<p style="padding-left: 40px;"><span style="font-family: , , 'segoe ui' , 'roboto' , 'oxygen' , 'ubuntu' , 'cantarell' , 'open sans' , 'helvetica neue' , sans-serif; font-size: medium;">3. Consider using temporary views instead of a derived table if you really need to use subselect in from clause consider creating a view using it in the query and dropping it after the query was executed.</span></p>

<h4 id="Sample:-Complex-Derived-Table"><span style="text-decoration: underline;">Sample: Complex Derived Table</span></h4>
<p><span style="font-size: 14px;">I have a problem executing the query in multiple lines. Just copy and paste the code in the DBeaver.</span></p>

```mysql
use classicmodels;
SELECT 
    customerNumber,
    ROUND(SUM(quantityOrdered * priceEach)) sales,
    (CASE
        WHEN SUM(quantityOrdered * priceEach) < 10000 THEN 'Silver'
        WHEN SUM(quantityOrdered * priceEach) BETWEEN 10000 AND 100000 THEN 'Gold'
        WHEN SUM(quantityOrdered * priceEach) > 100000 THEN 'Platinum'
    END) customerGroup
FROM
    orderdetails
        INNER JOIN
    orders USING (orderNumber)
WHERE
    YEAR(shippedDate) = 2003
GROUP BY customerNumber;
```

<h4 id="Use-Case:-Derived-Table-Iteration"><span style="text-decoration: underline;"><strong>Use Case: Derived Table Iteration</strong></span></h4>
<p><span style="font-size: 12pt;">This is very useful in reports generating dates. For example if you have a sales data from this query.</span></p>

```mysql
select cast('2021-01-03' as date) as transdate, 155 amount union all
select cast('2021-01-05' as date) as transdate, 174 amount union all
select cast('2021-01-13' as date) as transdate, 380 amount union all
select cast('2021-01-17' as date) as transdate, 235 amount union all
select cast('2021-01-29' as date) as transdate, 715 amount
```

Some business use cases will look for those missing dates on the report. Dates with no sales should display zero. In order to do that, we need to create a fake data of dates for the month of January.

```mysql
select 	date_add('2021-01-01', interval (0 + i1.i + i2.i + i3.i + i4.i + i5.i + i6.i) day) as transdate
from 	(select 0 as i union select 1) as i1, 
	 	(select 0 as i union select 2) as i2, 
		(select 0 as i union select 4) as i3, 
		(select 0 as i union select 8) as i4, 
		(select 0 as i union select 16) as i5,
		(select 0 as i union select 32) as i6
```

Then we join the fake data of dates to the sales derived table.

```mysql
select		dt.transdate, ifnull(sales.amount,0) amount 
from 		(
			select 	date_add('2021-01-01', interval (0 + i1.i + i2.i + i3.i + i4.i + i5.i + i6.i) day) as transdate
			from 	(select 0 as i union select 1) as i1, 
				 	(select 0 as i union select 2) as i2, 
					(select 0 as i union select 4) as i3, 
					(select 0 as i union select 8) as i4, 
					(select 0 as i union select 16) as i5,
					(select 0 as i union select 32) as i6
			) dt
left join	(
			select cast('2021-01-03' as date) as transdate, 155 amount union all
			select cast('2021-01-05' as date) as transdate, 174 amount union all
			select cast('2021-01-13' as date) as transdate, 380 amount union all
			select cast('2021-01-17' as date) as transdate, 235 amount union all
			select cast('2021-01-29' as date) as transdate, 715 amount
			) as sales on sales.transdate=dt.transdate
```

<h2 id="5.-Sub-Query" style="font-family: 'helvetica neue' , 'helvetica' , 'arial' , sans-serif; color: #000000;">5. <span style="text-decoration: underline;">Sub Query</span></h2>
<p><span style="text-decoration: underline;"><a style="background-color: #ffeebb; color: #000000; font-family: 'arial' , 'helvetica' , 'verdana' , sans-serif; font-size: 12.48px;" name="CIHFHJHJ" shape="rect"></a></span></p>
<p style="margin-top: 8px; margin-bottom: 8px;">A MySQL subquery is a query nested within another query such as SELECT, INSERT, UPDATE or DELETE. In addition, a subquery can be nested inside another subquery.</p>
<p style="margin-top: 8px; margin-bottom: 8px;"><span style="font-size: 14px;">I have a problem executing the query in multiple lines. Just copy and paste the code in the DBeaver.</span></p>
<p style="margin-top: 8px; margin-bottom: 8px;"><span style="font-size: 14px;">The <span style="text-decoration: underline;"><strong><span style="color: #e03e2d; text-decoration: underline;"><em>Outer Query</em></span></strong></span> and Inner <span style="text-decoration: underline;"><span style="color: #e03e2d;"><strong>Query (Sub Query)</strong></span></span></span></p>

```mysql
select		e.*
from 		employees e 
where 		e.officeCode
	in 		(
			select  o2.officeCode 
			from 	offices o2 
			where	o2.country = 'USA'
			)
```

<h4 id="In-and-Not-In-Operators" style="font-family: 'helvetica neue' , 'helvetica' , 'arial' , sans-serif; color: #000000;"><span style="text-decoration: underline;">In and Not In Operators</span></h4>
<p><span style="font-family: 'helveticaneue'; font-size: 12pt;">If a subquery returns more than one value, you can use other operators such as IN or NOT IN operation in the WHERE clause. This query will get the customers who have orders.</span></p>

```mysql
SELECT  	customerName
FROM		customers
WHERE   	customerNumber 
	NOT IN 	(SELECT customerNumber FROM orders GROUP BY customerNumber);
```

<h4 id="Exists-and-Not-Exists" style="font-family: 'helvetica neue' , 'helvetica' , 'arial' , sans-serif; color: #000000;"><span style="text-decoration: underline;">Exists and Not Exists</span></h4>
<p><span style="font-family: 'helveticaneue'; font-size: 12pt;">When a subquery is used with the EXISTS or NOT EXISTS operator, a subquery returns a boolean.&nbsp;</span><span style="font-family: , , 'segoe ui' , 'roboto' , 'oxygen' , 'ubuntu' , 'cantarell' , 'open sans' , 'helvetica neue' , sans-serif; font-size: medium;">The following query finds sales orders whose total values are greater than 60K.</span></p>

```mysql
SELECT    	 orderNumber
    		,SUM(priceEach * quantityOrdered) total
FROM		orderdetails
INNER JOIN  orders USING (orderNumber)
GROUP BY 	orderNumber
HAVING 		SUM(priceEach * quantityOrdered) > 60000;
```

<h4><span style="font-size: 12pt;">CREATE TRIGGER</span></h4>
<p><span style="font-size: 12pt;">Syntax</span></p>

```mysql
CREATE TRIGGER trigger_name
{BEFORE | AFTER} {INSERT | UPDATE| DELETE }
ON table_name FOR EACH ROW
BEGIN
trigger_body;
END;
```

<p style="padding-left: 40px;"><span style="font-size: 12pt;">1. </span><span style="font-family: , , 'segoe ui' , 'roboto' , 'oxygen' , 'ubuntu' , 'cantarell' , 'open sans' , 'helvetica neue' , sans-serif; font-size: medium;">First, specify the name of the trigger that you want to create after the&nbsp;</span><span style="font-size: 12pt;"><code style="font-family: monospace , monospace; margin: 0px; padding-top: 0px; padding-right: 5px; padding-left: 5px; border: none; background-color: #fff6ea;">CREATE TRIGGER</code></span><span style="font-family: , , 'segoe ui' , 'roboto' , 'oxygen' , 'ubuntu' , 'cantarell' , 'open sans' , 'helvetica neue' , sans-serif; font-size: medium;">&nbsp;</span><span style="font-family: , , 'segoe ui' , 'roboto' , 'oxygen' , 'ubuntu' , 'cantarell' , 'open sans' , 'helvetica neue' , sans-serif; font-size: medium;">keywords. Note that the trigger name must be unique within a database.</span></p>
<p style="padding-left: 40px;"><span style="font-family: , , 'segoe ui' , 'roboto' , 'oxygen' , 'ubuntu' , 'cantarell' , 'open sans' , 'helvetica neue' , sans-serif; font-size: medium;">2. </span><span style="font-family: , , 'segoe ui' , 'roboto' , 'oxygen' , 'ubuntu' , 'cantarell' , 'open sans' , 'helvetica neue' , sans-serif; font-size: medium;">Next, specify the trigger action time which can be either&nbsp;</span><code style="font-family: monospace , monospace; font-size: 0.9rem; margin: 0px; padding-top: 0px; padding-right: 5px; padding-left: 5px; border: none; background-color: #fff6ea;">BEFORE</code><span style="font-family: , , 'segoe ui' , 'roboto' , 'oxygen' , 'ubuntu' , 'cantarell' , 'open sans' , 'helvetica neue' , sans-serif; font-size: medium;">&nbsp;</span><span style="font-family: , , 'segoe ui' , 'roboto' , 'oxygen' , 'ubuntu' , 'cantarell' , 'open sans' , 'helvetica neue' , sans-serif; font-size: medium;">or</span><span style="font-family: , , 'segoe ui' , 'roboto' , 'oxygen' , 'ubuntu' , 'cantarell' , 'open sans' , 'helvetica neue' , sans-serif; font-size: medium;">&nbsp;</span><code style="font-family: monospace , monospace; font-size: 0.9rem; margin: 0px; padding-top: 0px; padding-right: 5px; padding-left: 5px; border: none; background-color: #fff6ea;">AFTER</code><span style="font-family: , , 'segoe ui' , 'roboto' , 'oxygen' , 'ubuntu' , 'cantarell' , 'open sans' , 'helvetica neue' , sans-serif; font-size: medium;">&nbsp;</span><span style="font-family: , , 'segoe ui' , 'roboto' , 'oxygen' , 'ubuntu' , 'cantarell' , 'open sans' , 'helvetica neue' , sans-serif; font-size: medium;">which indicates that the trigger is invoked before or after each row is modified.</span></p>
<p style="padding-left: 40px;"><span style="font-family: , , 'segoe ui' , 'roboto' , 'oxygen' , 'ubuntu' , 'cantarell' , 'open sans' , 'helvetica neue' , sans-serif; font-size: medium;">3. </span><span style="font-family: , , 'segoe ui' , 'roboto' , 'oxygen' , 'ubuntu' , 'cantarell' , 'open sans' , 'helvetica neue' , sans-serif; font-size: medium;">Then, specify the operation that activates the trigger, which can be INSERT, UPDATE, or DELETE.</span></p>
<p style="padding-left: 40px;"><span style="font-family: , , 'segoe ui' , 'roboto' , 'oxygen' , 'ubuntu' , 'cantarell' , 'open sans' , 'helvetica neue' , sans-serif; font-size: medium;">4. </span><span style="font-family: , , 'segoe ui' , 'roboto' , 'oxygen' , 'ubuntu' , 'cantarell' , 'open sans' , 'helvetica neue' , sans-serif; font-size: medium;">After that, specify the name of the table to which the trigger belongs after the ON</span><span style="font-family: , , 'segoe ui' , 'roboto' , 'oxygen' , 'ubuntu' , 'cantarell' , 'open sans' , 'helvetica neue' , sans-serif; font-size: medium;">&nbsp;</span><span style="font-family: , , 'segoe ui' , 'roboto' , 'oxygen' , 'ubuntu' , 'cantarell' , 'open sans' , 'helvetica neue' , sans-serif; font-size: medium;">keyword.</span></p>
<p style="padding-left: 40px;"><span style="font-family: , , 'segoe ui' , 'roboto' , 'oxygen' , 'ubuntu' , 'cantarell' , 'open sans' , 'helvetica neue' , sans-serif; font-size: medium;">5. </span><span style="font-family: , , 'segoe ui' , 'roboto' , 'oxygen' , 'ubuntu' , 'cantarell' , 'open sans' , 'helvetica neue' , sans-serif; font-size: medium;">Finally, specify the statement to execute when the trigger activates. If you want to execute multiple statements, you use the BEGIN END </span><span style="font-family: , , 'segoe ui' , 'roboto' , 'oxygen' , 'ubuntu' , 'cantarell' , 'open sans' , 'helvetica neue' , sans-serif; font-size: medium;">compound statement.</span></p>
<p><span style="font-family: , , 'segoe ui' , 'roboto' , 'oxygen' , 'ubuntu' , 'cantarell' , 'open sans' , 'helvetica neue' , sans-serif; font-size: medium;">First, we need to create a table employee_audit.</span></p>
<p>&nbsp;</p>

```mysql
CREATE TABLE IF NOT EXISTS employees_audit (
    id INT AUTO_INCREMENT PRIMARY KEY,
    employeeNumber INT NOT NULL,
    lastname VARCHAR(50) NOT NULL,
    changedat DATETIME DEFAULT NULL,
    action VARCHAR(50) DEFAULT NULL
);
```

<p><span style="font-size: 12pt;">Next, create a BEFORE UPDATE trigger that is invoked before a change is made to the employees table.</span></p>

```mysql
CREATE TRIGGER before_employee_update 
    BEFORE UPDATE ON employees
    FOR EACH ROW 
 INSERT INTO employees_audit
 SET action = 'update',
     employeeNumber = OLD.employeeNumber,
     lastname = OLD.lastname,
     changedat = NOW();
```

<p><span style="font-size: 12pt;">Let's check if the triggers works...</span></p>

```mysql
-- lets update the employees table
update employees set lastName = 'Murphyy' where employeeNumber=1002
-- then verify the employees_audit table.
select * from employees_audit
```

The IF EXISTS</span><span style="font-family: , , 'segoe ui' , 'roboto' , 'oxygen' , 'ubuntu' , 'cantarell' , 'open sans' , 'helvetica neue' , sans-serif; font-size: medium;">&nbsp;</span><span style="font-family: , , 'segoe ui' , 'roboto' , 'oxygen' , 'ubuntu' , 'cantarell' , 'open sans' , 'helvetica neue' , sans-serif; font-size: medium;">clause is optional.</span></p>
<p style="padding-left: 40px;">&nbsp;</p>
<p><span style="font-family: , , 'segoe ui' , 'roboto' , 'oxygen' , 'ubuntu' , 'cantarell' , 'open sans' , 'helvetica neue' , sans-serif; font-size: medium;">Create a clone of the before_employee_update trigger and drop it afterward.</span></p>

```mysql
CREATE TRIGGER before_employee_update_clone
    BEFORE UPDATE ON employees
    FOR EACH ROW 
 INSERT INTO employees_audit
 SET action = 'update',
     employeeNumber = OLD.employeeNumber,
     lastname = OLD.lastname,
     changedat = NOW();
```

In addition, you can use&nbsp;</span><a style="background-color: #ffffff; color: var(--link-color); transition: all 0.1s ease-in-out 0s; font-family: -apple-system, system-ui, 'Segoe UI', Roboto, Oxygen, Ubuntu, Cantarell, 'Open Sans', 'Helvetica Neue', sans-serif; font-size: medium;" href="https://www.mysqltutorial.org/mysql-delete-statement.aspx">DELETE</a><span style="font-family: -apple-system, system-ui, 'Segoe UI', Roboto, Oxygen, Ubuntu, Cantarell, 'Open Sans', 'Helvetica Neue', sans-serif; font-size: medium;">&nbsp;statement to remove rows of the underlying table through the view.&nbsp;&nbsp;</span><span style="font-family: -apple-system, system-ui, 'Segoe UI', Roboto, Oxygen, Ubuntu, Cantarell, 'Open Sans', 'Helvetica Neue', sans-serif; font-size: medium;">However, to create an updatable&nbsp;</span><a style="background-color: #ffffff; color: var(--link-color); transition: all 0.1s ease-in-out 0s; font-family: -apple-system, system-ui, 'Segoe UI', Roboto, Oxygen, Ubuntu, Cantarell, 'Open Sans', 'Helvetica Neue', sans-serif; font-size: medium;" title="Introduction to Database View" href="https://www.mysqltutorial.org/introduction-sql-views.aspx">view</a><span style="font-family: -apple-system, system-ui, 'Segoe UI', Roboto, Oxygen, Ubuntu, Cantarell, 'Open Sans', 'Helvetica Neue', sans-serif; font-size: medium;">, the&nbsp;</span><a style="background-color: #ffffff; color: var(--link-color); transition: all 0.1s ease-in-out 0s; font-family: -apple-system, system-ui, 'Segoe UI', Roboto, Oxygen, Ubuntu, Cantarell, 'Open Sans', 'Helvetica Neue', sans-serif; font-size: medium;" title="MySQL Select" href="https://www.mysqltutorial.org/mysql-select-statement-query-data.aspx">SELECT statement</a><span style="font-family: -apple-system, system-ui, 'Segoe UI', Roboto, Oxygen, Ubuntu, Cantarell, 'Open Sans', 'Helvetica Neue', sans-serif; font-size: medium;"> that defines the view&nbsp;must not contain any of the following elements:</span></p>

<h4 id="VIEW-WITH-CHECK-OPTION-CLAUSE">VIEW WITH CHECK OPTION CLAUSE</h4>
<p style="margin-bottom: 26px; padding: 0px; font-family: , , 'segoe ui' , 'roboto' , 'oxygen' , 'ubuntu' , 'cantarell' , 'open sans' , 'helvetica neue' , sans-serif; font-size: medium;">Sometimes, you create a view to reveal the partial data of a table. However, a simple view is updatable therefore it is possible to udpate&nbsp;data which is not visible through the view. This update&nbsp;makes the view inconsistent. To ensure the consistency of the view, you use the&nbsp;<span style="font-size: 12pt;"><code style="font-family: monospace , monospace; margin: 0px; padding-top: 0px; padding-right: 5px; padding-left: 5px; border: none; background-color: #fff6ea;">WITH CHECK OPTION</code></span> clause when you create or modify the view. The&nbsp;<span style="font-size: 12pt;"><code style="font-family: monospace , monospace; margin: 0px; padding-top: 0px; padding-right: 5px; padding-left: 5px; border: none; background-color: #fff6ea;">WITH CHECK OPTION</code></span>&nbsp;is an optional clause of the&nbsp;<span style="font-size: 12pt;"><code style="font-family: monospace , monospace; margin: 0px; padding-top: 0px; padding-right: 5px; padding-left: 5px; border: none; background-color: #fff6ea;">CREATE VIEW</code></span>&nbsp;statement. The&nbsp;<span style="font-size: 12pt;"><code style="font-family: monospace , monospace; margin: 0px; padding-top: 0px; padding-right: 5px; padding-left: 5px; border: none; background-color: #fff6ea;">WITH CHECK OPTION</code></span> prevents a view from updating or inserting rows that are not visible through it. In other words, whenever you update or insert a row of the base tables through a view, MySQL ensures that the insert or update operation is conformed with the definition of the view.</p>
<p style="margin-bottom: 26px; padding: 0px; font-family: , , 'segoe ui' , 'roboto' , 'oxygen' , 'ubuntu' , 'cantarell' , 'open sans' , 'helvetica neue' , sans-serif; font-size: medium;"><span style="font-size: 12pt;">You just put the&nbsp;</span><span class="hljs-keyword" style="color: #333333; font-weight: bold; font-family: monospace , monospace; font-size: 14.4px; white-space: pre;">WITH</span> <span class="hljs-keyword" style="color: #333333; font-weight: bold; font-family: monospace , monospace; font-size: 14.4px; white-space: pre;">CHECK</span> <span class="hljs-keyword" style="color: #333333; font-weight: bold; font-family: monospace , monospace; font-size: 14.4px; white-space: pre;">OPTION</span><span style="color: #333333; font-family: monospace , monospace; font-size: 14.4px; white-space: pre; background-color: #f8f8f8;">;</span><span style="font-size: 12pt;"> after the select statement of the view.</span></p>

<p><span style="font-size: 12pt;">Let's put it in action... Create a view and name it vw_vp_employees that filters the employee's jobtitle having the word 'VP'.</span></p>

```mysql
CREATE OR REPLACE VIEW vw_vp_employees AS
SELECT 	    employeeNumber
		    ,lastname
		    ,firstname
		    ,jobtitle
		    ,extension
		    ,email
		    ,officeCode
		    ,reportsTo
FROM		employees
WHERE	    jobTitle LIKE '%VP%'
WITH CHECK OPTION;
```

Next, query data from the vw_vp_employees view using the following SELECT statement:

```mysql
select * from vw_vp_employees;
```

<p><span style="font-size: 12pt;">Then, insert a row into the employees table through the vw_vp_employees view.</span></p>

```mysql
INSERT INTO vw_vp_employees(employeeNumber,firstName,lastName,jobTitle,extension,email,officeCode,reportsTo) 
SELECT 1703,'Lily','Bush','IT Manager','x9111','lilybush@classicmodelcars.com',1,1002;
```

<h3>8. STORED FUNCTION</h3>
<p><span style="font-family: -apple-system, system-ui, 'Segoe UI', Roboto, Oxygen, Ubuntu, Cantarell, 'Open Sans', 'Helvetica Neue', sans-serif; font-size: medium;">A stored function is a special kind stored program that returns a single value. Typically, you use stored functions to encapsulate common formulas or business rules that are reusable among SQL statements or stored programs.</span></p>

<p><span style="font-size: 12pt;">Let's put it in action...&nbsp;<span style="font-family: -apple-system, system-ui, 'Segoe UI', Roboto, Oxygen, Ubuntu, Cantarell, 'Open Sans', 'Helvetica Neue', sans-serif;">The following&nbsp;</span><code style="font-family: monospace, monospace; margin: 0px; padding-top: 0px; padding-right: 5px; padding-left: 5px; border: none; background-color: #fff6ea; white-space: normal;">CREATE FUNCTION</code><span style="font-family: -apple-system, system-ui, 'Segoe UI', Roboto, Oxygen, Ubuntu, Cantarell, 'Open Sans', 'Helvetica Neue', sans-serif;"> statement creates a function that returns the customer level based on credit</span></span></p>

```mysql
CREATE FUNCTION fn_customer_level(
	credit DECIMAL(10,2)
) 
RETURNS VARCHAR(20)
DETERMINISTIC
BEGIN
    DECLARE customerLevel VARCHAR(20);

    IF credit > 50000 THEN
		SET customerLevel = 'PLATINUM';
    ELSEIF (credit >= 50000 AND 
			credit <= 10000) THEN
        SET customerLevel = 'GOLD';
    ELSEIF credit < 10000 THEN
        SET customerLevel = 'SILVER';
    END IF;
	-- return the customer level
	RETURN (customerLevel);
END
```

<p><span style="font-size: 12pt;">Let's check the function created.</span></p>

```mysql
SHOW FUNCTION STATUS WHERE db = 'classicmodels';
```

<p><span style="font-size: 12pt;">Then, we call the function....</span></p>

```mysql
SHOW FUNCTION STATUS WHERE db = 'classicmodels';
SELECT 		customerName, creditLimit, fn_customer_level(creditLimit)
FROM    	customers
ORDER BY    customerName;
```

<h2 id="9.-TRANSACTIONS">9. TRANSACTIONS</h2>
<p><span style="font-family: , , 'segoe ui' , 'roboto' , 'oxygen' , 'ubuntu' , 'cantarell' , 'open sans' , 'helvetica neue' , sans-serif; font-size: medium;">To understand what a transaction in MySQL is, let&rsquo;s take a look at an example of adding a new sales order</span><span style="font-family: , , 'segoe ui' , 'roboto' , 'oxygen' , 'ubuntu' , 'cantarell' , 'open sans' , 'helvetica neue' , sans-serif; font-size: medium;">. The steps of adding a sales order are as described as follows:</span></p>
<p style="padding-left: 40px;"><span style="font-family: , , 'segoe ui' , 'roboto' , 'oxygen' , 'ubuntu' , 'cantarell' , 'open sans' , 'helvetica neue' , sans-serif; font-size: medium;">1. </span><span style="font-family: , , 'segoe ui' , 'roboto' , 'oxygen' , 'ubuntu' , 'cantarell' , 'open sans' , 'helvetica neue' , sans-serif; font-size: medium;">First, query</span><span style="font-family: , , 'segoe ui' , 'roboto' , 'oxygen' , 'ubuntu' , 'cantarell' , 'open sans' , 'helvetica neue' , sans-serif; font-size: medium;">&nbsp;</span><span style="font-family: , , 'segoe ui' , 'roboto' , 'oxygen' , 'ubuntu' , 'cantarell' , 'open sans' , 'helvetica neue' , sans-serif; font-size: medium;">the&nbsp;latest sales order number from the&nbsp;</span><span style="font-size: 12pt;"><code style="font-family: monospace , monospace; margin: 0px; padding-top: 0px; padding-right: 5px; padding-left: 5px; border: none; background-color: #fff6ea;">orders</code></span><span style="font-family: , , 'segoe ui' , 'roboto' , 'oxygen' , 'ubuntu' , 'cantarell' , 'open sans' , 'helvetica neue' , sans-serif; font-size: medium;">&nbsp;</span><span style="font-family: , , 'segoe ui' , 'roboto' , 'oxygen' , 'ubuntu' , 'cantarell' , 'open sans' , 'helvetica neue' , sans-serif; font-size: medium;">table and use the next sales order number as the new sales order number.</span></p>
<p style="padding-left: 40px;"><span style="font-family: , , 'segoe ui' , 'roboto' , 'oxygen' , 'ubuntu' , 'cantarell' , 'open sans' , 'helvetica neue' , sans-serif; font-size: medium;">2. </span><span style="font-family: , , 'segoe ui' , 'roboto' , 'oxygen' , 'ubuntu' , 'cantarell' , 'open sans' , 'helvetica neue' , sans-serif; font-size: medium;">Next, insert</span><span style="font-family: , , 'segoe ui' , 'roboto' , 'oxygen' , 'ubuntu' , 'cantarell' , 'open sans' , 'helvetica neue' , sans-serif; font-size: medium;">&nbsp;</span><span style="font-family: , , 'segoe ui' , 'roboto' , 'oxygen' , 'ubuntu' , 'cantarell' , 'open sans' , 'helvetica neue' , sans-serif; font-size: medium;">a new sales order into the&nbsp;</span><span style="font-size: 12pt;"><code style="font-family: monospace , monospace; margin: 0px; padding-top: 0px; padding-right: 5px; padding-left: 5px; border: none; background-color: #fff6ea;">orders</code></span><span style="font-family: , , 'segoe ui' , 'roboto' , 'oxygen' , 'ubuntu' , 'cantarell' , 'open sans' , 'helvetica neue' , sans-serif; font-size: medium;">&nbsp;</span><span style="font-family: , , 'segoe ui' , 'roboto' , 'oxygen' , 'ubuntu' , 'cantarell' , 'open sans' , 'helvetica neue' , sans-serif; font-size: medium;">table.</span></p>
<p style="padding-left: 40px;"><span style="font-family: , , 'segoe ui' , 'roboto' , 'oxygen' , 'ubuntu' , 'cantarell' , 'open sans' , 'helvetica neue' , sans-serif; font-size: medium;">3. </span><span style="font-family: , , 'segoe ui' , 'roboto' , 'oxygen' , 'ubuntu' , 'cantarell' , 'open sans' , 'helvetica neue' , sans-serif; font-size: medium;">Then, get the newly inserted sales order number.</span></p>
<p style="padding-left: 40px;"><span style="font-family: , , 'segoe ui' , 'roboto' , 'oxygen' , 'ubuntu' , 'cantarell' , 'open sans' , 'helvetica neue' , sans-serif; font-size: medium;">4. </span><span style="font-family: , , 'segoe ui' , 'roboto' , 'oxygen' , 'ubuntu' , 'cantarell' , 'open sans' , 'helvetica neue' , sans-serif; font-size: medium;">After that, insert the new sales order items into the&nbsp;</span><span style="font-size: 12pt;"><code style="font-family: monospace , monospace; margin: 0px; padding-top: 0px; padding-right: 5px; padding-left: 5px; border: none; background-color: #fff6ea;">orderdetails</code></span><span style="font-family: , , 'segoe ui' , 'roboto' , 'oxygen' , 'ubuntu' , 'cantarell' , 'open sans' , 'helvetica neue' , sans-serif; font-size: medium;">&nbsp;</span><span style="font-family: , , 'segoe ui' , 'roboto' , 'oxygen' , 'ubuntu' , 'cantarell' , 'open sans' , 'helvetica neue' , sans-serif; font-size: medium;">table with the sales order number.</span></p>
<p style="padding-left: 40px;"><span style="font-family: , , 'segoe ui' , 'roboto' , 'oxygen' , 'ubuntu' , 'cantarell' , 'open sans' , 'helvetica neue' , sans-serif; font-size: medium;">5. </span><span style="font-size: 12pt;"><span style="font-family: , , 'segoe ui' , 'roboto' , 'oxygen' , 'ubuntu' , 'cantarell' , 'open sans' , 'helvetica neue' , sans-serif;">Finally, select data from both&nbsp;</span><code style="font-family: monospace , monospace; margin: 0px; padding-top: 0px; padding-right: 5px; padding-left: 5px; border: none; background-color: #fff6ea;">orders</code><span style="font-family: , , 'segoe ui' , 'roboto' , 'oxygen' , 'ubuntu' , 'cantarell' , 'open sans' , 'helvetica neue' , sans-serif;">&nbsp;</span><span style="font-family: , , 'segoe ui' , 'roboto' , 'oxygen' , 'ubuntu' , 'cantarell' , 'open sans' , 'helvetica neue' , sans-serif;">and</span><span style="font-family: , , 'segoe ui' , 'roboto' , 'oxygen' , 'ubuntu' , 'cantarell' , 'open sans' , 'helvetica neue' , sans-serif;">&nbsp;</span><code style="font-family: monospace , monospace; margin: 0px; padding-top: 0px; padding-right: 5px; padding-left: 5px; border: none; background-color: #fff6ea;">orderdetails</code><span style="font-family: , , 'segoe ui' , 'roboto' , 'oxygen' , 'ubuntu' , 'cantarell' , 'open sans' , 'helvetica neue' , sans-serif;">&nbsp;</span><span style="font-family: , , 'segoe ui' , 'roboto' , 'oxygen' , 'ubuntu' , 'cantarell' , 'open sans' , 'helvetica neue' , sans-serif;">tables to confirm the changes.</span></span></p>
<p><span style="font-family: , , 'segoe ui' , 'roboto' , 'oxygen' , 'ubuntu' , 'cantarell' , 'open sans' , 'helvetica neue' , sans-serif; font-size: medium;">Now, imagine what would happen to the sales order data if one or more steps above fail?&nbsp;</span><span style="font-family: , , 'segoe ui' , 'roboto' , 'oxygen' , 'ubuntu' , 'cantarell' , 'open sans' , 'helvetica neue' , sans-serif; font-size: medium;">That is why the transaction processing comes to the rescue. MySQL transaction allows you to execute a set of MySQL operations to ensure that the database never contains the result of partial operations. In a set of operations, if one of them fails, the rollback occurs to restore the database to its original state. If no error occurs, the entire set of statements is committed to the database.</span></p>

<p><span style="font-size: 12pt;">Let's put it in action...</span></p>

```mysql
-- 1. start a new transaction
START TRANSACTION;

-- 2. Get the latest order number
SELECT @orderNumber:=MAX(orderNUmber)+1 FROM orders;

-- 3. insert a new order for customer 145
INSERT INTO orders(orderNumber, orderDate, requiredDate, shippedDate, status, customerNumber)
VALUES(@orderNumber, '2005-05-31',  '2005-06-10', '2005-06-11', 'In Process', 145);
        
-- 4. Insert order line items
INSERT INTO orderdetails(orderNumber,  productCode, quantityOrdered, priceEach, orderLineNumber)
VALUES(@orderNumber,'S18_1749', 30, '136', 1),
      (@orderNumber,'S18_2248', 50, '55.09', 2); 
      
-- 5. commit changes    
COMMIT;
```

<h2 id="9.-TEMPORARY-TABLE" style="font-family: 'helvetica neue' , 'helvetica' , 'arial' , sans-serif; color: #000000;">9. TEMPORARY TABLE</h2>
<p><span style="font-family: , , 'segoe ui' , 'roboto' , 'oxygen' , 'ubuntu' , 'cantarell' , 'open sans' , 'helvetica neue' , sans-serif; font-size: medium;">A temporary table is a special type of table that allows you to store a temporary result set, which you can reuse several times in a single session.&nbsp;</span></p>
<p><strong><span style="font-size: 14pt;">FEATURES</span></strong></p>
<p style="padding-left: 40px;"><span style="font-size: 12pt;">1. </span><span style="font-family: , , 'segoe ui' , 'roboto' , 'oxygen' , 'ubuntu' , 'cantarell' , 'open sans' , 'helvetica neue' , sans-serif; font-size: medium;">MySQL removes the temporary table automatically when the session ends or the connection is terminated. Of course, you can use the&nbsp; DROP TABLE</span><span style="font-family: , , 'segoe ui' , 'roboto' , 'oxygen' , 'ubuntu' , 'cantarell' , 'open sans' , 'helvetica neue' , sans-serif; font-size: medium;">&nbsp;</span><span style="font-family: , , 'segoe ui' , 'roboto' , 'oxygen' , 'ubuntu' , 'cantarell' , 'open sans' , 'helvetica neue' , sans-serif; font-size: medium;">statement to remove a temporary table explicitly when you are no longer use it.</span></p>
<p style="padding-left: 40px;"><span style="font-family: , , 'segoe ui' , 'roboto' , 'oxygen' , 'ubuntu' , 'cantarell' , 'open sans' , 'helvetica neue' , sans-serif; font-size: medium;">2. </span><span style="font-family: , , 'segoe ui' , 'roboto' , 'oxygen' , 'ubuntu' , 'cantarell' , 'open sans' , 'helvetica neue' , sans-serif; font-size: medium;">A temporary table is only available and accessible to the client that creates it. Different clients can create temporary tables with the same name without causing errors because only the client that creates the temporary table can see it. However, in the same session, two temporary tables cannot share the same name.</span></p>
<p>Syntax</p>

```mysql
CREATE TEMPORARY TABLE table_name(
   column_1_definition,
   column_2_definition,
   ...,
   table_constraints
);
```

<p><span style="font-size: 12pt;">Let's put in action... Create a temporary table and name it tmp_credit that stores customer's credit.</span></p>

```mysql
CREATE TEMPORARY TABLE tmp_credits(
    customerNumber INT PRIMARY KEY,
    creditLimit DEC(10,2)
);
```

<p><span style="font-family: -apple-system, system-ui, 'Segoe UI', Roboto, Oxygen, Ubuntu, Cantarell, 'Open Sans', 'Helvetica Neue', sans-serif; font-size: medium;">Then, insert rows from the&nbsp;</span><span style="font-size: 12pt;"><code style="font-family: monospace, monospace; margin: 0px; padding-top: 0px; padding-right: 5px; padding-left: 5px; border: none; background-color: #fff6ea; white-space: normal;">customers</code></span><span style="font-family: -apple-system, system-ui, 'Segoe UI', Roboto, Oxygen, Ubuntu, Cantarell, 'Open Sans', 'Helvetica Neue', sans-serif; font-size: medium;"> table into the temporary table tmp_credits</span><span style="font-family: -apple-system, system-ui, 'Segoe UI', Roboto, Oxygen, Ubuntu, Cantarell, 'Open Sans', 'Helvetica Neue', sans-serif; font-size: medium;">:</span></p>

```mysql
INSERT INTO tmp_credits(customerNumber,creditLimit)
SELECT customerNumber, creditLimit
FROM customers
WHERE creditLimit > 0;
```

<h3 id="10.-STORED-PROCEDURE">10. STORED PROCEDURE</h3>
<p><span style="font-size: 12pt;">Advantages</span></p>
<p style="padding-left: 40px;"><span style="font-size: 12pt;">1.&nbsp;</span><span style="font-family: -apple-system, system-ui, 'Segoe UI', Roboto, Oxygen, Ubuntu, Cantarell, 'Open Sans', 'Helvetica Neue', sans-serif; font-size: medium;">Stored procedures help reduce the network traffic between applications and MySQL Server. Because instead of sending multiple lengthy SQL statements, applications have to send only the name and parameters of stored procedures.</span></p>
<p style="padding-left: 40px;"><span style="font-family: -apple-system, system-ui, 'Segoe UI', Roboto, Oxygen, Ubuntu, Cantarell, 'Open Sans', 'Helvetica Neue', sans-serif; font-size: medium;">2.&nbsp;</span><span style="font-family: -apple-system, system-ui, 'Segoe UI', Roboto, Oxygen, Ubuntu, Cantarell, 'Open Sans', 'Helvetica Neue', sans-serif; font-size: medium;">You can use the stored procedures to implement business logic that is reusable by multiple applications. The stored procedures help reduce the efforts of duplicating the same logic in many applications and make your database more consistent.</span></p>
<p style="padding-left: 40px;"><span style="font-family: -apple-system, system-ui, 'Segoe UI', Roboto, Oxygen, Ubuntu, Cantarell, 'Open Sans', 'Helvetica Neue', sans-serif; font-size: medium;">3.&nbsp;</span><span style="font-family: -apple-system, system-ui, 'Segoe UI', Roboto, Oxygen, Ubuntu, Cantarell, 'Open Sans', 'Helvetica Neue', sans-serif; font-size: medium;">The database administrator can grant appropriate privileges to applications that only access specific stored procedures without giving any privileges on the underlying tables.</span></p>
<p><span style="font-family: -apple-system, system-ui, 'Segoe UI', Roboto, Oxygen, Ubuntu, Cantarell, 'Open Sans', 'Helvetica Neue', sans-serif; font-size: medium;">Disadvantage</span></p>
<p style="padding-left: 40px;"><span style="font-family: -apple-system, system-ui, 'Segoe UI', Roboto, Oxygen, Ubuntu, Cantarell, 'Open Sans', 'Helvetica Neue', sans-serif; font-size: medium;">1. Resource usage.&nbsp;</span><span style="font-family: -apple-system, system-ui, 'Segoe UI', Roboto, Oxygen, Ubuntu, Cantarell, 'Open Sans', 'Helvetica Neue', sans-serif; font-size: medium;">If you use many stored procedures, the memory usage of every connection will increase substantially.</span></p>
<p>Syntax</p>

```mysql
CREATE PROCEDURE procedure_name(parameter_list)
BEGIN
   statements;
END //
```

<p style="padding-left: 40px;"><span style="font-size: 12pt;">1. </span><span style="font-family: , , 'segoe ui' , 'roboto' , 'oxygen' , 'ubuntu' , 'cantarell' , 'open sans' , 'helvetica neue' , sans-serif; font-size: medium;">First, specify the name of the stored procedure that you want to create after the&nbsp;</span><span style="font-size: 12pt;"><code style="font-family: monospace , monospace; margin: 0px; padding-top: 0px; padding-right: 5px; padding-left: 5px; border: none; background-color: #fff6ea;">CREATE PROCEDURE</code><span style="font-family: , , 'segoe ui' , 'roboto' , 'oxygen' , 'ubuntu' , 'cantarell' , 'open sans' , 'helvetica neue' , sans-serif;">&nbsp;</span></span><span style="font-family: , , 'segoe ui' , 'roboto' , 'oxygen' , 'ubuntu' , 'cantarell' , 'open sans' , 'helvetica neue' , sans-serif; font-size: medium;">keywords.</span></p>
<p style="padding-left: 40px;"><span style="font-family: , , 'segoe ui' , 'roboto' , 'oxygen' , 'ubuntu' , 'cantarell' , 'open sans' , 'helvetica neue' , sans-serif; font-size: medium;">2. </span><span style="font-family: , , 'segoe ui' , 'roboto' , 'oxygen' , 'ubuntu' , 'cantarell' , 'open sans' , 'helvetica neue' , sans-serif; font-size: medium;">Second, specify a list of comma-separated parameters for the stored procedure in parentheses after the procedure name.</span></p>
<p style="padding-left: 40px;"><span style="font-family: , , 'segoe ui' , 'roboto' , 'oxygen' , 'ubuntu' , 'cantarell' , 'open sans' , 'helvetica neue' , sans-serif; font-size: medium;">3. </span><span style="font-family: , , 'segoe ui' , 'roboto' , 'oxygen' , 'ubuntu' , 'cantarell' , 'open sans' , 'helvetica neue' , sans-serif; font-size: medium;">Third, write the code between the&nbsp;</span><span style="font-size: 12pt;"><code style="font-family: monospace , monospace; margin: 0px; padding-top: 0px; padding-right: 5px; padding-left: 5px; border: none; background-color: #fff6ea;">BEGIN END</code><span style="font-family: , , 'segoe ui' , 'roboto' , 'oxygen' , 'ubuntu' , 'cantarell' , 'open sans' , 'helvetica neue' , sans-serif;">&nbsp;</span></span><span style="font-family: , , 'segoe ui' , 'roboto' , 'oxygen' , 'ubuntu' , 'cantarell' , 'open sans' , 'helvetica neue' , sans-serif; font-size: medium;">block. The above example just has a simple</span><span style="font-family: , , 'segoe ui' , 'roboto' , 'oxygen' , 'ubuntu' , 'cantarell' , 'open sans' , 'helvetica neue' , sans-serif; font-size: medium;"> SELECT</span><span style="font-family: , , 'segoe ui' , 'roboto' , 'oxygen' , 'ubuntu' , 'cantarell' , 'open sans' , 'helvetica neue' , sans-serif; font-size: medium;">&nbsp;</span><span style="font-family: , , 'segoe ui' , 'roboto' , 'oxygen' , 'ubuntu' , 'cantarell' , 'open sans' , 'helvetica neue' , sans-serif; font-size: medium;">statement. After the</span><span style="font-family: , , 'segoe ui' , 'roboto' , 'oxygen' , 'ubuntu' , 'cantarell' , 'open sans' , 'helvetica neue' , sans-serif; font-size: medium;">&nbsp;</span><span style="font-size: 12pt;"><code style="font-family: monospace , monospace; margin: 0px; padding-top: 0px; padding-right: 5px; padding-left: 5px; border: none; background-color: #fff6ea;">END</code></span><span style="font-family: , , 'segoe ui' , 'roboto' , 'oxygen' , 'ubuntu' , 'cantarell' , 'open sans' , 'helvetica neue' , sans-serif; font-size: medium;">&nbsp;</span><span style="font-family: , , 'segoe ui' , 'roboto' , 'oxygen' , 'ubuntu' , 'cantarell' , 'open sans' , 'helvetica neue' , sans-serif; font-size: medium;">keyword, you place the delimiter character to end the procedure statement.</span></p>

<p><span style="font-size: 12pt;">Let's create a stored procedure and name it sp_sales_order. Then, copy the code from the transaction example. Modify some values in the insert statement and put them inside the begin-end block.</span></p>

```mysql
CREATE PROCEDURE sp_sales_order()
BEGIN
-- 1. start a new transaction
START TRANSACTION;

-- 2. Get the latest order number
SELECT @orderNumber:=MAX(orderNUmber)+1 FROM orders;

-- 3. insert a new order for customer 145
INSERT INTO orders(orderNumber, orderDate, requiredDate, shippedDate, status, customerNumber)
VALUES(@orderNumber, '2005-05-30',  '2005-06-11', '2005-06-13', 'In Process', 145);
        
-- 4. Insert order line items
INSERT INTO orderdetails(orderNumber,  productCode, quantityOrdered, priceEach, orderLineNumber)
VALUES(@orderNumber,'S18_1749', 30, '121.25', 1),
      (@orderNumber,'S18_2248', 50, '175.10', 2); 
      
-- 5. commit changes    
COMMIT;
END 
```

<p><span style="font-size: 12pt;">Let's put it in action...</span></p>

```mysql
call sp_sales_order();
```
